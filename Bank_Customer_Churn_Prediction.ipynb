Here in above column names 'RowNumber', 'CustomerId', 'Surname' these columns are not necessery for our prediction it wont affect statistic data. So drop these column fields.

````python
df1=df.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1) # This DROP method will drop columns mentioned in bracket
df1 # Displaying new dataframe with droped columns (Without head() and tail() method it will only show top 5 and bottom 5 records.)
````

As we can see in above dataset some categorical values which are repeatative and machine only understands ordinal values. Finding unique values from categorical  values.

````python
df1.columns
````

````python
print(df1['Geography'].unique())
print(df1['Gender'].unique())
print(df1['NumOfProducts'].unique())
print(df1['HasCrCard'].unique())
print(df1['IsActiveMember'].unique())
````

Converting these repeatative categorical values to numerical, so using lable encoding and avoiding dummies.

````python
df2=pd.get_dummies(df1,drop_first=True)
df2
````

Now in above df2 column 'Exited' is Y dependent factor and others are X independent factor.

````python
df2['Exited'].value_counts() #This VALUE_COUNT method gives distinct count of each nos.
````

Here we get to know 1 is exited customers and 0 is staying customers. Which shows this is classification problem.

Dividing data into Dependent and Independent variables.

````python
x=df2.drop('Exited',axis=1) #independent factors
y=df2['Exited'] #dependent factors
````

````python
x
````

````python
y
````

Further spliting x and y in train, test models.

````python
from sklearn.model_selection import train_test_split
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42,stratify=y)
````

We are scaling dataset to standardise scaling of whole dataset.

````python
from sklearn.preprocessing import StandardScaler
````

````python
sc=StandardScaler()
````

````python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
````

In our case dependent variable is classification based so we are using classification algorithm.

````python
from sklearn.linear_model import LogisticRegression
````

````python
log = LogisticRegression() #creating instance of algorithm
````

````python
log.fit(x_train,y_train)
````

````python
y_pred1=log.predict(x_test)
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(y_test,y_pred1)
````

````python
from sklearn.metrics import precision_score,recall_score,f1_score
````

````python
precision_score(y_test,y_pred1)*100
````

````python
recall_score(y_test,y_pred1)
````

````python
f1_score(y_test,y_pred1)
````

As accuracy score value and other  F1_score, precision_score and recall_score showing large difference. It shows theat data is imbalanced. So we are performing OverSampling as Undersampling causes to loose valuable data using SMOTE analysis.

````python
from imblearn.over_sampling import SMOTE
````

````python
x_res,y_res=SMOTE().fit_resample(x,y)
````

````python
y_res.value_counts()
````

Now we can see there is no difference in customer Exited an retained. Again performing test train split on new values.

````python
X_train,X_test,Y_train,Y_test=train_test_split(x_res,y_res,test_size=0.2,random_state=42)
````

````python
y_pred2=log.predict(X_test)
````

````python
accuracy_score(Y_test,y_pred2)
````

````python
precision_score(Y_test,y_pred2)
````

````python
recall_score(Y_test,y_pred2)
````

````python
f1_score(Y_test,y_pred2)
````
